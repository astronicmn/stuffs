```python
# Importando módulos:
from scipy.stats import poisson 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Poisson é definido como f(k) = \exp(-\mu) \frac{\mu^k}{k!}

x = np.arange(0, 50, 1) # Definindo um intervalo para o plot
y = poisson.pmf(x, mu=12, loc=0) # mu: localização de µ / loc: localização do parâmetro
  
plt.plot(x, y) 
plt.show() 
```

```python
def poi(k, l):
  f = ((l**k) * np.exp(-l)) / np.math.factorial(k) # Poisson
  return f

f = poi(10, x)

plt.plot(x, f, c="mediumpurple")
plt.plot(x, y, c="magenta")
plt.show()
```
