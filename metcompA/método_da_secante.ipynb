Criando um código que receba a função \begin{equation}
f(x) =-x^2+6x+10
\end{equation} dois pontos iniciais (a e b) e a tolerância; retornando, assim, o valor da raiz. 

```python
def f(x): # apenas a definição da função escolhida
  x = -x**2+6*x+10 
  return x
```

```python
def secante(f,a,b): # função secante
  tole = 1e-5 
  contador1 = 200
  contador2 = 0
  while True: # loop que realiza os cálculos da função e repete o processo com o
              # novo ponto (x) e um dos pontos iniciais (a ou b),
              # escolhendo aquele que estiver mais próximo do novo ponto.
    x = a - ((a - b)/(f(a)-f(b)))*f(a)
    erro = abs((x - a)/x)
    b = a
    a = x
    contador2 += 1
    if contador2 > contador1: # meios para parar o loop (contador ou tolerância)
      break
    if erro < tole:
      break
  return float('%.2f'%a)
```

```python
secante(f, 1, 7)
```
