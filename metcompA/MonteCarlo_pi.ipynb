# Estimativa de $\pi$ pelo método Monte Carlo

```python
import numpy as np
import matplotlib.pyplot as plt 
import random 
```

```python
# Crio vetores x e y, contendo N números aleatórios


N = 10_000

x= np.random.random(N)
y= np.random.random(N)

z_in  =np.where(np.sqrt(x**2 + y**2) <= 1.0)  # Verifico quais pares caem dentro do círculo
z_out =np.where(np.sqrt(x**2 + y**2) > 1.0)

## As linhas a seguir preparam os dados para produzir o gráfico da próxima célula

x_in =[]         #Crio listas com as coordenadas dos pontos interiores ao círculo
y_in =[]

x_out =[]        #Crio listas com as coordenadas dos pontos interiores ao círculo
y_out =[]

for i in z_in:
    x_in.append(x[i])   
    y_in.append(y[i])

for i in z_out:
    x_out.append(x[i])
    y_out.append(y[i])


x_in = np.asarray(x_in)   # converto as listas em np.arrays
y_in = np.asarray(y_in)

x_out = np.asarray(x_out) # converto as listas em np.arrays
y_out = np.asarray(y_out)
```

```python
xx = np.linspace(0,1,50)
yy = np.sqrt(1-xx**2)     # produz gráfico da quadrante circular
```

```python
plt.figure(figsize=(5,5))
plt.plot(xx,yy, lw=3,c='blue', ls='--')  # quadrante circular
plt.scatter(x_in,y_in, facecolor=None,edgecolor='g', alpha=0.1) # pontos interiores
plt.scatter(x_out,y_out, facecolor=None,edgecolor='r', alpha=0.1) # pontos exteriores
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

## Determino valor de $\pi$

```python
pi = 4.0 * len(z_in[0])/N
print(pi)
```

## Determinando o valor de $\pi$ para vários valores de N

```python
def estima_pi(N):  # Função que estima pi para N pares de números aleaórios
    x = np.random.random(N)
    y = np.random.random(N)
    z_in = np.where(np.sqrt(x**2 + y**2) <= 1.0)  # Verifico quais pares caem dentro do círculo
    return 4.0 * len(z_in[0])/N
```

```python
NN = [1_000, 10_000, 100_000, 1_000_000, 10_000_000, 25_000_000,50_000_000, 80_000_000, 100_000_000]
```

```python
Pi = []
for N in NN:
    Pi.append(estima_pi(N))
```

```python
plt.plot(NN, Pi, '-o')
plt.axhline(y = np.pi, ls='--', c='r')
plt.xlabel('N')
plt.ylabel(r'$\pi$', size=20)
plt.grid()
plt.show()
```
