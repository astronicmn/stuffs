Série de Taylor com telescopagem + funcão fatorial recursiva.

```python
def fatorial(n):
    fat=1
    for i in range(n):
        fat = fat * (i+1)
    return fat    
```

```python
def taylor_exp(x):
    exp = 1.0
    diff = 1.0e5
    eps = 1.0e-6
    prev =1.0
    i = 1.0
    
    while (diff >=  eps):
        ant = exp
        #print(i, exp, diff)
        corr = x/i * prev
        prev = corr
        exp = exp + corr
        i = i + 1
        diff = abs(exp - ant)
    return exp    
```

```python
def taylor_mod(x):
    exp = 1.0
    diff = 1.0e5
    eps = 1.0e-6
    prev =1.0
    i = 1
    
    while (diff >=  eps):
        ant = exp
        #print(i, exp, diff)
        exp = exp + x**i/fatorial(i)        
        i = i + 1
        diff = abs(exp - ant)
    return exp    
```

```python
%timeit taylor_exp(1.0)
```

```python
%timeit taylor_mod(1.0)
```
