```python
def distrib(N):
  amostra = []
  for i in range(1000):
    y = np.random.choice(x, N)
    amostra.append(y.mean())
  z = np.asarray(amostra)
  return z.std()
```

```python
NN = np.asarray([2, 12, 20, 50, 100, 250, 500, 750, 1000])
```

```python
estats = []
for N in NN:
  estats.append(distrib(N))
```
