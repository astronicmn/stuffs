## Crie uma função em python que implemente o algoritmo babilônio. Compare os resultados com a função sqrt() do módulo math.

```python
import math
from math import sqrt
import cmath
```

```python
def babilonico(N, xy = 1):

  xy1 = N
  E = 0.0001

  while(xy1 - xy) > E:
    
    aux1 = N / xy
    aux2 = aux1 + xy
    xy1 = aux2 / 2
    xy = xy + 1

  print(xy1)
```

```python
babilonico(100)
```

```python
babilonico(66)
```

```python
sqrt(100)
```

```python
sqrt(66)
```
