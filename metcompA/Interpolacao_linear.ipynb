### Interpolação polinomial

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def seno(x, h):
  return np.sin(x), np.sin(x+h)

x = np.linspace(0, 2* np.pi, 25)
x
```

```python
plt.plot(x, np.sin(x), '*', ms=12)
```

```python
def interpolar(x1, x2, y1, y2):
  y = y1 + (x2 - x) * (y2 - y1/(x2 - x1))
  return y
```

```python
h = (x[1]-x[0])/2.0
sen_interp = []
x_interp = []

for xx in x:
  x1 = xx
  x2 = xx + h
  y1, y2 = seno(x1, h)
  z = interpolar(x1, x2, y1, y2)
  sen_interp.append(y2)
  x_interp.append(x2)
```

```python
plt.plot(x, np.sin(x), '*', ms = 10)
plt.plot(x_interp, sen_interp, 'o', ms = 8)
plt.show()
```
