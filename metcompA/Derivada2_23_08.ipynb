```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivada(f,x_0,dx,N=3):
    x_1 = x_0 - dx
    x_2 = x_0 + dx
    deriv = (f(x_2) - f(x_1))/(2*dx)
    der_avan = (f(x_2) - f(x_0))/(dx)
    der_ret = (f(x_0) - f(x_1))/(dx)
    return deriv

def der_avan(f,x_0,dx,N=3):
    x_1 = x_0 - dx
    x_2 = x_0 + dx
    deriv = (f(x_2) - f(x_1))/(2*dx)
    der_avan = (f(x_2) - f(x_0))/(dx)
    der_ret = (f(x_0) - f(x_1))/(dx)
    return der_avan

def der_ret(f,x_0,dx,N=3):
    x_1 = x_0 - dx
    x_2 = x_0 + dx
    deriv = (f(x_2) - f(x_1))/(2*dx)
    der_avan = (f(x_2) - f(x_0))/(dx)
    der_ret = (f(x_0) - f(x_1))/(dx)
    return der_ret
```

### Derivada simétrica sobre 3 pontos:
$$  hf^{'} = \frac{1}{2} \left( f_{1} - f_{-1} \right)  $$

```python
f = np.sin
x = np.linspace(0,2*np.pi,50)
```

```python
dx = 1.0e-2
y1 = derivada(f,x,dx)
y2 = der_avan(f,x,dx)
y3 = der_ret(f,x,dx)
```

```python
plt.plot(x, y1, lw=3, c='darkblue')
plt.plot(x, y2, lw=3, c='darkcyan')
plt.plot(x, y3, lw=3, c='darkgray')
plt.scatter(x, np.cos(x), marker='o', c='red')
plt.grid()
plt.show()
```

```python
plt.scatter(x, y1 - np.cos(x), marker='o', c='red')
plt.grid()
plt.show()
```

```python
def derivada_5p(f,x_0,dx):
    x_1 = x_0 - 2*dx
    x_2 = x_0 - dx
    x_3 = x_0 + dx
    x_4 = x_0 + 2*dx
    
    
    deriv = (f(x_1) - 8* f(x_2) + 8*f(x_3) -f(x_4))/(12*dx)
    
    return deriv
```

```python
dx = 1.0
y1 = derivada(f,x,dx)
y2 = derivada_5p(f,x,dx)
```

```python
plt.plot(x, y1, label ='3 pontos')
plt.plot(x, y2, label ='5 pontos')
plt.scatter(x,np.cos(x), label ='exato')
plt.grid()
plt.legend()
plt.show()
```

## Entendendo a evolução dos erros com o valor de dx

```python
h =[10, 5, 2, 1, 0.5, 0.2, 0.1, 0.05, 0.02, 0.01, 0.005, 0.002, 0.001, 0.0005, 0.0002, 0.0001, 0.00005, 0.00002, 0.00001 ]
```

```python
y1=[]
y2=[]
y3=[]

x=1.0
print(np.cos(x))

for dx in h:
    y1.append(derivada(f,x,dx))
    y2.append(der_avan(f,x,dx))
    y3.append(der_ret(f,x,dx))
```

```python
plt.semilogx(h,y1 - np.cos(x),label='3 pontos')
plt.semilogx(h,y2 - np.cos(x),label='x e x + dx')
plt.semilogx(h,y3 - np.cos(x),label='x e x - dx') 
plt.xlim([1e-3,10])
plt.xlabel('dx')
plt.ylabel(r'Erro ($\epsilon$)')
plt.grid()
plt.legend()
plt.show()
```
