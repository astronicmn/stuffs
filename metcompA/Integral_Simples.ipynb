# Exerc√≠cio:
\begin{equation}
\int_{0}^{1}(3x^2 + 4x - 2)dx
\end{equation}

```python
N = []
for i in range (50):
  N.append(2+i*2)
NL = len(N)
```

```python
def f(x):
  f = 3*x**2 + 4.0*x - 2
  return f
```

```python
x = np.linspace(0, 4, NL)
```

```python
def trapezio(N, a, b):
  h = (b-a)/N
  I = (f(b) + f(a)) * h/2.0
  for i in range(1,N):
    x = a + h * i
    I = I + f(x) * h
  return I
```

```python
a = 0.0
b = 4.0
N = 100

I = trapezio(N, a, b)
print(I)
```

```python
intervalos = [10, 20, 40, 80, 160, 320, 640, 1280]
I=[]
for N in intervalos:
  I.append(trapezio(N, a, b))
plt.plot(intervalos, I, "k")
plt.grid()
```
