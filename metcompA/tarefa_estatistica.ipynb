## Escreva uma função em python que calcule a média, o desvio padrão e o erro da
## média de uma lista de valores numéricos. Lembrando que:

##$ x_m = \frac{Σ^N_ix_i}{N}, \sigma^2 = \frac{Σ(x_i-x_m)^2}{N-1}, δx_m = \frac{σ}{\sqrt{N}}$ 


```python
import random as rnd
x = rnd.random()
x
```

```python
N = 10
list = []
for i in range(N):
  list.append(rnd.random())
list
```

```python
from math import sqrt
S = 0.0
S2 = 0.0

N = len(list)
for i in range(N):
  S = S + list[i]
  S2 = S2 + (list[i])**2

S = S/N
S2 = S2/N - S**2
sigma = sqrt(S2)
delta = sigma/sqrt(N)
```

```python
print('Valor médio:', S)
print('Variância:', S2)
print('Erro na média:', delta)
```
