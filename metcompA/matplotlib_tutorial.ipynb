```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use(['science', 'notebook', 'grid']) #Aplico os estilos. Os estilos disponíveis estão descritos em https://github.com/garrettj403/SciencePlots
```

# Exemplos de plots simples 

```python
# bom para uma inspeção rápida dos dados
x = np.linspace(0,15,30)
y = np.sin(x) + 0.1*np.random.randn(len(x)) # função seno com ruído adicionado
y2 = np.cos(x)
```

```python
plt.plot(x,y) # Grafico de linha simples
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y, '*--', color='purple', lw=2, ms=10) # utilizando símbolos e linhas tracejadas. 
                                                  # lw: largura da linha (Line Width)
                                                  # ms: tamanho do símbolo (Marker Size)
```

#### Ajustando o tamanho da figura

```python
plt.figure(figsize=(8,3))
plt.plot(x,y)
```

#### Legendas para eixos

```python
plt.figure(figsize=(8,3))
plt.plot(x,y)
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]')
plt.title('Voltage in 3rd Electrode')
plt.show()
```

#### Legenda do gráfico

```python
plt.figure(figsize=(8,3))
plt.plot(x,y, label='Component 1')
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]')
plt.title('Voltage in 3rd Electrode')
#plt.legend(loc='lower right', fontsize=18)
plt.legend( fontsize=18)
plt.show()
```

#### Colocando vários gráficos na mesma figura:

```python
x2 = np.linspace(0, 15, 100)
y2 = np.sin(x2)
```

```python
plt.figure(figsize=(8,3))
plt.plot(x,y, 'o', label='Data', zorder=1)
plt.plot(x2,y2, label='Fit', zorder=2)
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]')
plt.title('Voltage in 3rd Electrode')
plt.legend(loc='lower right', fontsize=12)
plt.show()
```

#### Controlando os limites $x$ e $y$

```python
plt.figure(figsize=(8,3))
plt.plot(x,y, 'o', label='Data', zorder=1)
plt.plot(x2,y2, label='Fit', zorder = 2) # zorder: linhas sobre símbolos
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]')
plt.title('Voltage in 3rd Electrode')
plt.legend(loc='upper right', fontsize=12, ncol=2) # Número de colunas da legenda
plt.ylim(top=2)
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.plot(x,y, 'o', label='Data', zorder=2)# zorder: símbolos sobre linhas
plt.plot(x2,y2, label='Fit', zorder = 1) 
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]')
plt.title('Voltage in 3rd Electrode')
plt.legend(loc='upper right', fontsize=12, ncol=2) # Número de colunas da legenda
plt.ylim(top=2)
plt.show()
```

# Histogramas

```python
res = np.random.randn(1000)*0.2 + 0.4  #randn devolve números aleatórios distribuidos como uma gaussiana. 
```

```python
plt.figure(figsize=(8,3))
plt.hist(res) # comando para preparar histograma
plt.show()
```

```python
# Os mesmos dados, com mais intervalos:
plt.figure(figsize=(8,3))
plt.hist(res, bins=30) #30 intervalos
plt.show()
```

```python
# Os mesmos dados, com o histograma normalizado (i.e, integral = 1)
plt.figure(figsize=(8,3))
plt.hist(res, bins=30, density=True) # density=True - histograma normalizado
plt.show()
```

```python
#Vários estilos de intervalos, para comparar histogramas na mesma figura #histtype
res2 = np.random.randn(1000)*0.2 + 0.4 # gera novo set de dados.
```

```python
plt.figure(figsize=(8,3))
plt.hist(res, bins=30, density=True, histtype='step')
plt.hist(res2, bins=30, density=True, histtype='step')
plt.xlabel('$\hat{E}-E$', fontsize=20)
plt.ylabel('Frequency')
plt.show()
```

```python
# O mesmo gráfico, sem a especificação de  hysttype fica difícil de analisar:
plt.figure(figsize=(8,3))
plt.hist(res, bins=30, density=True)
plt.hist(res2, bins=30, density=True)
plt.xlabel('$\hat{E}-E$', fontsize=20)
plt.ylabel('Frequency')
plt.show()
```

## Gráficos mais complexos exigem uma API mais avançada:

```python
fig, ax = plt.subplots(1, 1, figsize=(12,4))
ax.plot(x, y, 'o--', color='r', lw=0.4, ms=3)
ax.text(0.1, 0.1, 'text here', transform=ax.transAxes) ## coordenadas relativas aos Axes da figura
                                                       ## esses vão de (0,0) no canto inferior esquerdo 
                                                       ## a (1,1) no canto superior direito
                                                       ## Compare com próximo gráfico
ax.set_xlabel('Time [s]')
ax.set_ylabel(r'$\frac{d}{dx} f(x)$', fontsize=15)     ## Latex para gerar o texto do label vertical
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,4))
ax.plot(x, y, 'o--', color='b', lw=0.6, ms=4)
ax.text(0.9, -0.9, 'text here', transform=ax.transData)## coordenadas relativas aos dados da figura,
                                                       ## controlados por xlim e ylim 
ax.set_xlabel('Time [s]')
ax.set_ylabel(r'$\frac{d}{dx} f(x)$', fontsize=15)
plt.show()
```

## Múltiplos Axes em uma única figura:

```python
fig, axes = plt.subplots(3, 2, figsize=(8,8))
ax = axes[0][0]
ax.plot(x, y, 'o--', color='r', lw=0.4, ms=3)
ax.text(0.1, 0.1, 'text here', transform=ax.transAxes)
ax = axes[1][1]
ax.hist(res, bins=30, density=True, histtype='step', label='C1')
ax.hist(res2, bins=30, density=True, histtype='step', label='C2')
ax.set_xlabel('$\hat{E}-E$')
ax.legend()
fig.tight_layout()
fig.suptitle('Title of All Plots', y=1.03, fontsize=25)
plt.show()
```

```python
## Ajustando o tamanho dos rótulos nos eixos
plt.style.use(['science', 'notebook', 'grid'])
fig, ax = plt.subplots(1, 1, figsize=(8,3))
ax.plot(x, y, 'o--', color='r', lw=1, ms=5)
ax.text(0.1, 0.1, 'text here', transform=ax.transAxes)
ax.set_xlabel('Time [s]')
ax.set_ylabel(r'$\frac{d}{dx} f(x)$', fontsize=15)
ax.tick_params(axis='both', which='major', labelsize=10) #Ajuste dos tamanhos dos labels
ax.grid('minor', color='green')
plt.show()
```
