# Lançamento de projéteis

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
t = np.linspace(0,20,400);
g=9.81
```

```python
@interact(thet=(0.0,90.0))
def trajectory(thet=45.0):
    v_0 = 30.0
    theta = thet *(2*np.pi/360.00)

    v0_y = v_0 * np.sin(theta)
    v0_x = v_0 * np.cos(theta)
    x_0 = 0.0
    y_0 = 0.0
    x = x_0 + v0_x * t
    y = y_0 + v0_y * t  - (1./2.)*g*t**2
    plt.plot(x,y)
    plt.ylim([0,50])
    plt.xlim([0,100])
    plt.grid()
    plt.show()
```
