```python
d = {"first": "Albert", 
     "last": "Einstein", 
     "birthday": [1879, 3, 14]}

print("Keys:")
for key in d.keys():
    print(key)

print("\n======\n")

print("Values:")
for value in d.values():
    print(value)

print("\n======\n")

print("Items:")
for key, value in d.items():
    print(key, value)
```

```python
from math import sin

def sin_inv_x(x):
    if x == 0.0:
        result = 0.0
    else:
        result = sin(1.0/x)
    return result
```

```python
def power(base, x):
    """Computes base^x. Both base and x should be integers,
    floats, or another numeric type.
    """
    return base**x
```

```python
def line(x, a=1.0, b=0.0):
    return a*x + b
```

```python
line(42)  # no keyword args, returns 1*42 + 0
```

```python
line(42, 2)         # a=2, returns 84
```

```python
line(42, b=10)      # b=10, returns 52
```

```python
line(42, b=10, a=2) # returns 94 
```

```python
line(42, a=2, b=10) # also returns 94 
```

```python
def minimum(*args):
    """Takes any number of arguments!"""
    m = args[0]
    for x in args[1:]:
        if x < m:
          m = x
    return m 
```

```python
minimum(6, 42)
```

```python
data = [65, 42, 2, 8]
minimum(*data)
```
