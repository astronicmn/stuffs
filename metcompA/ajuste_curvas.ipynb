## Ajuste de curvas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## y = a * x + b
```

```python
def reta(x, a=1, b=0):
  y = a * x + b 
  return y
```

```python
x = np.linspace(-2, 2, 20)
y = reta(x, 2, 3)
N = len(y)
dados = y + np.random.rand(N)*0.5

plt.plot(x, y)
plt.plot(x, dados, 'o',  color='purple')
plt.grid()
```

```python
z = np.polyfit(x, dados, deg=1, cov=True) #cov = matriz de covari√¢ncia
z #erro
```
