```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit #numba é um compilador 
```

```python
Npoints = 21
x = np.linspace(0,2,Npoints) #1001 espaços/intervalos, daí sabemos o valor de h
                          
x = x * np.pi

y = np.sin(x) #calculando seno

yy = np.cos(x) #calculando cosseno

plt.plot(x, y)
```

```python
@njit
def deriv(y, x): #definindo a função 
  resultado = []
  h = x[1]-x[0]
  n_points = np.shape(x)[0]

  for i in range(1, n_points-1):
    deriv = y[i+1] - y[i-1]/(2*h)
    resultado.append(deriv)

  return np.asarray(resultado)
```

```python
%timeit der = deriv(y,x)
```

```python
plt.scatter(x[1:-1], der)
plt.plot(x[1:-1], yy[1:-1], c = 'red')
plt.grid()
```
